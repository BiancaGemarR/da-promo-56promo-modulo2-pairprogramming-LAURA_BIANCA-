```python
#1. importar librerías
import mysql.connector
from mysql.connector import errorcode
import pandas as pd
```

```python
#2. crear conexión northwind y error si algo no sale bien 
try:
  cnx = mysql.connector.connect(user='root', password='AlumnaAdalab',
                              host='127.0.0.1',
                              database='northwind')
  print("Acces ok")
except mysql.connector.Error as err:
    if err.errno == errorcode.ER_ACCESS_DENIED_ERROR:
        print("Something is wrong with your user name or password")
    else:
        print(err)
else:
  cnx.close()
```

```python
#3. Usando "information_schema" en la conexión a base de datos, 
# extraer los nombres de las columnas de la tabla "customers" de la BBDD "northwind
cnx = mysql.connector.connect(user='root', password='AlumnaAdalab',
                              host='127.0.0.1',
                              database='northwind')
mycursor = cnx.cursor()
mycursor.execute("""SELECT COLUMN_NAME, DATA_TYPE, IS_NULLABLE, COLUMN_DEFAULT
                FROM information_schema.COLUMNS
                WHERE TABLE_SCHEMA = 'northwind' AND TABLE_NAME = 'customers'""")
mycursor.description
myresult = mycursor.fetchall()
df = pd.DataFrame(myresult)
```

```python
#4. Seleccionar todo de la tabla "customers" 
# donde "city" sea "London" o "Seattle". Almacenar los resultados en una variable.
mycursor = cnx.cursor()
sql = """SELECT * FROM customers
        WHERE city = 'London' or city = 'Seattle'"""
mycursor.execute(sql)
myresult = mycursor.fetchall()
df = pd.DataFrame(myresult)
```
